```python
df_heart = pd.get_dummies(df_heart_master, columns=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'], drop_first=True)
```

```python
df_heart.head()
```

```python
X_heart = df_heart.drop(columns=['HeartDisease'])
y_heart = df_heart['HeartDisease']
print(X_heart)
print(y_heart)
```

```python
X_diabetes = df_diabetes_master.drop(columns=['Diabetes_binary'])
y_diabetes = df_diabetes_master['Diabetes_binary']
print(X_diabetes)
print(y_diabetes)
```

```python
# training and evaluating on the heart dataset

X_train, X_test, y_train, y_test = train_test_split(X_heart, y_heart,random_state=0)
X_train_vals = X_train.values
X_test_vals = X_test.values
y_train_vals = y_train.values
y_test_vals = y_test.values


clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X_train_vals,y_train_vals)

y_pred = clf.predict(X_test_vals)
print('confusion matrix')
print(confusion_matrix(y_test_vals, y_pred))
print('Accuracy score')
print(accuracy_score(y_test_vals,y_pred))
print('recall/sensitivity score')
print(recall_score(y_test_vals,y_pred))
# tree.plot_tree(clf)
```

```python
# training and evaluating on the heart dataset

X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(X_diabetes, y_diabetes,random_state=0)
X_train_diabetes_vals = X_train_diabetes.values
X_test_diabetes_vals = X_test_diabetes.values
y_train_diabetes_vals = y_train_diabetes.values
y_test_diabetes_vals = y_test_diabetes.values


clf_diabetes = tree.DecisionTreeClassifier(max_depth=3)
clf_diabetes = clf_diabetes.fit(X_train_diabetes_vals, y_train_diabetes_vals)

y_pred_diabetes = clf_diabetes.predict(X_test_diabetes_vals)
print('confusion matrix')
print(confusion_matrix(y_test_diabetes_vals, y_pred_diabetes))
print('Accuracy score')
print(accuracy_score(y_test_diabetes_vals, y_pred_diabetes))
print('recall/sensitivity score')
print(recall_score(y_test_diabetes_vals, y_pred_diabetes))
#tree.plot_tree(clf)
#r = export_text(clf_diabetes, feature_names=list(X_diabetes.columns))
#print(r)
```
